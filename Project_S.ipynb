## Analisis kolom

### Subtask:
Melakukan analisis mendalam pada setiap kolom, termasuk tipe data, nilai yang hilang, nilai unik, dan distribusi data.


**Reasoning**:
Iterate through each column to analyze its type, missing values, unique values, and distribution, including attempts to convert 'object' type columns to numeric after cleaning.



```python
for col in df.columns:
    print(f"Column: {col}")
    print(f"Data type: {df[col].dtype}")
    missing_values = df[col].isnull().sum()
    print(f"Missing values: {missing_values}")

    unique_values = df[col].unique()
    if len(unique_values) < 20:
        print(f"Unique values: {unique_values}")
    else:
        print(f"Number of unique values: {len(unique_values)}")

    if df[col].dtype == 'object':
        # Attempt to clean and convert to numeric
        try:
            # Remove '.' and replace ',' with '.' for proper float conversion
            cleaned_col = df[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
            df[col] = pd.to_numeric(cleaned_col)
            print(f"Successfully converted '{col}' to numeric.")
            display(df[[col]].head())
            print(f"New data type for '{col}': {df[col].dtype}")
        except ValueError:
            print(f"Could not convert '{col}' to numeric. Manual inspection may be needed.")
            # For object columns that are not numeric, print value counts
            print(f"Value counts for '{col}':")
            display(df[col].value_counts())
    elif pd.api.types.is_numeric_dtype(df[col]):
        print(f"Descriptive statistics for '{col}':")
        display(df[col].describe())

    print("-" * 30)
```

## Visualisasi data

### Subtask:
Membuat grafik berwarna untuk memvisualisasikan hubungan antar kolom atau tren dalam data.


**Reasoning**:
Generate the visualizations based on the cleaned data.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert 'tahun' column to numeric if it's not already
if not pd.api.types.is_numeric_dtype(df['tahun']):
    df['tahun'] = pd.to_numeric(df['tahun'], errors='coerce')

plt.figure(figsize=(12, 6))
plt.title('Pendapatan Operasional vs. Tahun', fontsize=14)
sns.lineplot(x='tahun', y='pendapatan_operasional', data=df, marker='o', color='skyblue')
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Pendapatan Operasional', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
print("Visualisasi 1: Tren pendapatan operasional dari tahun ke tahun.")

plt.figure(figsize=(12, 6))
plt.title('Perbandingan Persediaan Awal dan Akhir', fontsize=14)
sns.barplot(x='tahun', y='persediaan_awal', data=df, label='Persediaan Awal', color='lightcoral')
sns.barplot(x='tahun', y='persediaan_akhir', data=df, label='Persediaan Akhir', color='lightgreen')
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Nilai Persediaan', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
print("Visualisasi 2: Perbandingan persediaan awal dan akhir setiap tahun.")


plt.figure(figsize=(12, 6))
plt.title('Hubungan antara Total HPP dan Laba Bersih', fontsize=14)
sns.scatterplot(x='total_hpp', y='laba_bersih_normal', data=df, hue='tahun', palette='viridis')
plt.xlabel('Total Harga Pokok Penjualan', fontsize=12)
plt.ylabel('Laba Bersih', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
print("Visualisasi 3: Hubungan antara total HPP dan laba bersih, diwarnai berdasarkan tahun.")
```
